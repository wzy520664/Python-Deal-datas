```python
#调节图片颜色
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from scipy import misc
```

```python
#引用mpimg
picture=mpimg.imread(r'C:\Users\Lenovo\python课设\孔雀.jpg')
```

```python
picture1=picture[:,:,0]
```

```python
#将图片转换 为蓝色
plt.imshow(picture1,cmap='Blues_r')
plt.show()
```

```python
#查看图片的不同方位
(h,w,c)=picture.shape
cx,cy=(w//2,h//2)
```

```python
#左上右上，左下右下
topleft=picture[0:cy,0:cx]
topright=picture[0:cy,cx:w]
bottomleft=picture[0:cy,0:cx]
bottomright=picture[0:cy,cx:w]
```

```python
#展示左上方
def show(picture):
    plt.imshow(picture)
    plt.axis('off')
show(topleft)
```

```python
#展示右上方
show(topright)
```

```python
#展示左下方
show(bottomleft)
```

```python
#展示右下方
show(bottomright)
```

```python
#对图片进行缩放处理
from PIL import Image
```

```python
#垂直翻转
picture2=picture[::-1,:,:]
plt.imshow(picture2)
```

```python
#水平翻转
picture3=picture[:,::-1,:]
plt.imshow(picture3)
```

```python
#180度翻转
picture4=picture[::-1,::-1,:]
plt.imshow(picture4)
```

```python
#宽度方向裁剪
w1=w//6
w2=w//3*2
picture5=picture[:,w1:w2,:]
plt.imshow(picture5)
```

```python
#高度方向裁剪
h=picture.shape[0]
w=picture.shape[1]
h1=h//2
h2=h
picture6=picture[h1:h2,:,:]
plt.imshow(picture6)
```

```python
#两个方向剪辑
picture7=picture[h1:h2,w1:w2,:]
plt.imshow(picture7)
```

```python
#调整亮度，亮度变为原来的0.4倍
picture8=picture*0.4
plt.imshow(picture8.astype('uint8'))
```

```python
#变亮，亮度变为之前的两倍
picture9=picture*2.0
picture9=np.clip(picture9,a_min=None,a_max=255)
plt.imshow(picture9.astype('uint8'))
```

```python
#间隔采样，图片尺寸减半。清晰度变差
picture10=picture[::2,::2,:]
plt.imshow(picture10)
```
